```python
import datetime
import numpy as np
import pandas as pd
import pandas_datareader.data as web

import seaborn as sns
sns.set_style("darkgrid")
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
stock = 'DPZ'

start = datetime.datetime(2018, 11, 1)
end = datetime.datetime(2019, 11, 1)




# start = datetime.datetime(2017, 5, 17)
# end = datetime.datetime(2019, 9, 19)
```

```python
def StockData(ticker, start_date, end_date):
    '''requires pandas datareader (as web) and date time'''
    df = web.DataReader(ticker, 'yahoo', start_date, end_date)
    return df
```

```python
dpz = StockData(stock, start, end)
dpz
```

```python
def PriceDate(stock_df):
    stock_df = stock_df.reset_index()
    stock_df['Price'] = stock_df['Adj Close']
    return stock_df[['Date', 'Price']]
```

```python
dpz = PriceDate(dpz)
dpz
```

```python
def drawdown(df):
    
    drawn_down_list = []
    peak_date = trough_date = df.Date[0]
    peak_price = trough_price = df.Price[0]
    continuing_peak = True
    
    for i in range(1, len(df)):
        date = df.Date[i]
        price = df.Price[i]
        
        if price < trough_price:
            trough_date = date
            trough_price = price
            
        if price > peak_price:
            if continuing_peak == False:
                drawn_down_list.append([peak_date, 
                                        trough_date,
                                        peak_price, 
                                        trough_price, 
                                        trough_date - peak_date, 
                                        peak_price -  trough_price, 
                                        trough_price / peak_price - 1]) 
            continuing_peak = True
            peak_date = date 
            trough_date = date
            peak_price = price
            trough_price = price
        else:
            continuing_peak = False

        if i == len(df) - 1:
            drawn_down_list.append([peak_date, 
                                    trough_date,
                                    peak_price,
                                    trough_price, 
                                    trough_date - peak_date, 
                                    peak_price -  trough_price, 
                                    trough_price / peak_price - 1]) 

    drawn_down_df = pd.DataFrame(drawn_down_list, columns=['Peak Date', 
                                                           'Trough Date',
                                                           'Peak Price',  
                                                           'Trough Price',
                                                           'Drawdown Days', 
                                                           'Drawdown Price', 
                                                           'Drawdown Pct'])

    return drawn_down_df
```

```python
drawdown(dpz)
```

```python
drawdown_df = drawdown(dpz)
```

```python
fig = plt.figure(figsize=(20,6))
sns.lineplot(dpz['Date'], dpz['Price'])

ax = fig.gca()

for i in range(len(drawdown_df)):
    ax.axvline(drawdown_df['Peak Date'][i], 
               color='green', 
               linestyle='--', 
               lw=1)

for i in range(len(drawdown_df)):
    ax.axvline(drawdown_df['Trough Date'][i], 
               color='red', 
               linestyle='--', 
               lw=1)


plt.title("Domino's Peaks and Troughs", fontsize=16)    
plt.xlabel('Date', fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.show()
```
